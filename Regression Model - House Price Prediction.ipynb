# How common factors affect the price of a house?

## Bedrooms

### What is the most common number of bedrooms in a house?

```python
df['bedrooms'].value_counts().plot(kind='bar', figsize=(15,5))
plt.title('Number of Bedrooms')
plt.xticks(rotation=0)
plt.xlabel('Bedrooms')
plt.ylabel('Count')
```

The price of the house increases as the number of bedrooms increases. The most expensive ones are those with 5 and 6 rooms. However, there are houses with 33 bedrooms that are cheaper than houses with 11 bedrooms. This is because the price of a house depends on other factors as well.

# Price prediction - ML model

## Pipeline

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_basement', 'floors', 'waterfront', 'view', 'yr_built', 'yr_renovated', 'zipcode']]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics
```

```python
model_pipeline = [LinearRegression(),
                  Lasso(alpha=0.01),
                  RandomForestRegressor(),
                  GradientBoostingRegressor()]
```

```python
model_list = ['Linear Regression', 'Lasso Regression', 'Random Forest', 'Gradient Boosting']
score_list = []
mae_list = []
mse_list = []
rmse_list = []
```

```python
for model in model_pipeline:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    score_list.append(model.score(X_test, y_test))
    mae_list.append(metrics.mean_absolute_error(y_test, predictions))
    mse_list.append(metrics.mean_squared_error(y_test, predictions))
    rmse_list.append(np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
results_df = pd.DataFrame({'Model': model_list, 'Score': score_list, 'MAE': mae_list, 'MSE': mse_list, 'RMSE': rmse_list})
```

```python
results_df
```

Based on the 'Score', the best model is the Random Forest, followed by the Gradient Boosting Regressor for the initial model.

## Fine-tuning the model

### Random Forest

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators': [100, 400, 800],
              'max_depth': [None, 2, 5],
              'min_samples_split': [2, 3, 7],
              'min_samples_leaf': [2, 5, 8],
              'n_jobs': [-1]}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print('Score:', grid.score(X_test, y_test))
print('MAE:', metrics.mean_absolute_error(y_test, grid_predictions))
print('MSE:', metrics.mean_squared_error(y_test, grid_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, grid_predictions)))
```

### Gradient Boosting

```python
clf = GradientBoostingRegressor(n_estimators = 500, max_depth = 5, min_samples_split = 2, learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(X_train, y_train)
```

```python
clf_predictions = clf.predict(X_test)
```

```python
print('Score:', clf.score(X_test, y_test))
print('MAE:', metrics.mean_absolute_error(y_test, clf_predictions))
print('MSE:', metrics.mean_squared_error(y_test, clf_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, clf_predictions)))
```

After fine-tuning the model, the best model is the Gradient Boosting Regressor with a score of 0.85 vs the Random Forest with a score of 0.81.
