```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error as mae
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('train.csv', sep = ';')
```

```python
df_test = pd.read_csv('test.csv', sep = ';')
```

```python
X = df[['feature1', 'feature2', 'feature3', 'feature4', 'feature5', 'feature6', 'feature7', 'feature8']]
y = df['target']
```

```python
for i in X:
    sns.regplot(x = df[i], y = y)
    plt.show()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y)
```

```python
rfc = RandomForestClassifier()

param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
```

```python
predictions_test = CV_rfc.predict(X_val)
mae(y_val, predictions_test)
```

```python
X_test = df_test[['feature1', 'feature2', 'feature3', 'feature4', 'feature5', 'feature6', 'feature7', 'feature8']]
```

```python
target = CV_rfc.predict(X_test)
```

```python
pd.DataFrame(dict(final_status=target))
pd.DataFrame({"final_status": target}).to_csv("predictions.csv", index=False)
```

```python
output_json = {"target": {str(i): int(pred) for i, pred in enumerate(target)}}
```

```python
import json
with open('result.json', 'w') as fp:
    json.dump(output_json, fp)
```
